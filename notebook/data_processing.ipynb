## Data processing - save into a txt file

```python
import pandas as pd 
df = pd.read_csv("../data.csv")
```

```python
df = df.dropna()
df.head()
```

```python
df['title'].tolist()[0:2]
```

```python
## save the titles into a txt file
with open("../data.txt", "a") as f:
    for title in df['title'].tolist():
        f.write(str(title))
        f.write("\n")
```
